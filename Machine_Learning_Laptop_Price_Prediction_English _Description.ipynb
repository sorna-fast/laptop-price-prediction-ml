````python
# Set the chart size
plt.figure(figsize=(12, 6))

# Draw a histogram with KDE line
sns.histplot(data=df, x='Price ($)', bins=30, kde=True, color='skyblue')

# Title and axes
plt.title('Laptop Price Distribution', fontsize=14)
plt.xlabel('price ($)', fontsize=12)
plt.ylabel('Number', fontsize=12)

# Show the chart
plt.show()
````

````python
def show_histplot(df: pd.DataFrame, numerical_features: list, target: str):
    df = df.copy()

    # Create price categories (low, medium, high)
    df['Price_Category'] = pd.cut(df[target],
    bins=[0, 1500, 3000, 11000],
    labels=['Low (0-1500$)', 'Medium (1500-3000$)', 'High (3000+$)'])

    # Choose a suitable color palette
    price_categories = ['Low (0-1500$)', 'Medium (1500-3000$)', 'High (3000+$)']
    colors = sns.color_palette('Set1', len(price_categories)) # Color set with 3 colors

    # Chart settings
    plt.figure(figsize=(20, 15))
    plt.suptitle('Comparison of feature distribution in different price ranges', fontsize=16, y=1.02) # 'Compare the distribution of features in different price ranges'

    # Draw a group histogram for each feature
    for i, feature in enumerate(numerical_features, 1):

        plt.subplot(2, 2, i)
        sns.histplot(data=df, x=feature, hue='Price_Category', bins=15, palette=colors, alpha=0.6)
        plt.title(f'Distribute {feature} by price', fontsize=12)
        plt.xlabel(feature, fontsize=10)
        plt.ylabel('Number', fontsize=10)

    # 📌 Add custom help
    patches = [mpatches.Patch(color=colors[i], label=price_categories[i]) for i in range(len(price_categories))]
    plt.legend(handles=patches, title='Price Category', loc='upper right', fontsize=12) # Show help in the upper right corner

    plt.tight_layout()
    plt.show()

# List of numerical features
numerical_features = ['RAM (GB)', 'Screen Size (inch)', 'Battery Life (hours)', 'Weight (kg)']
show_histplot(df=df, numerical_features=numerical_features, target='Price ($)')
````

````python
# TODO: Average Price by Processor

def show_mean_Processor_barplot(df:pd.DataFrame,target:str,y:str):
    df=df.copy()
    # Calculate average price for each processor
    processor_price = df.groupby('Processor')[target].mean().sort_values(ascending=False).reset_index()

    # Limit to top 10 processors (to avoid clutter)
    top_processors = processor_price.head(10)

    # Chart settings
    plt.figure(figsize=(14, 8))
    sns.barplot(data=top_processors, x=target, y=y, palette='viridis')

    # Add details to the chart
    plt.title('Average price of laptops by processor type (top 10 processors)', fontsize=14) # 'Average price of laptops by processor type (top 10 processors)'
    plt.xlabel('Average price ($)', fontsize=12)
    plt.ylabel('Processor Type', fontsize=12)
    plt.grid(axis='x', linestyle='--', alpha=0.7)

    # Show numeric value on bars
    for index, value in enumerate(top_processors[target]):
        plt.text(value + 50, index, f'{value:.0f}$', ha='left', va='center', fontsize=10)

    plt.tight_layout()
    plt.show()

show_mean_Processor_barplot(df=df,target='Price ($)',y="Processor")
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def show_avg_price_by_ram_type(df: pd.DataFrame, ram_type_col: str, price_col: str):
    # Remove NaN values ​​to avoid errors
    df = df.copy()

    # Calculate average price by RAM type
    ram_type_stats = df.groupby(ram_type_col)[price_col].mean().reset_index()

    # Rename columns for better readability
    ram_type_stats.columns = ['RAM Type', 'Mean Price ($)']

    # Sort by average price
    ram_type_stats = ram_type_stats.sort_values(by='Mean Price ($)', ascending=False)

    # Chart settings
    plt.figure(figsize=(12, 6))
    sns.barplot(data=ram_type_stats, x='RAM Type', y='Mean Price ($)', palette='coolwarm')

    # Appearance settings
    plt.xlabel('RAM Type', fontsize=12)
    plt.ylabel('Average Price ($)', fontsize=12)
    plt.title('Average Laptop Price by RAM Type', fontsize=14)
    plt.xticks(rotation=45) # Rotate RAM type name for better readability
    plt.grid(axis='y', linestyle='--', alpha=0.7) # Guide lines

    plt.show()

# Execute function to display chart
show_avg_price_by_ram_type(df, ram_type_col='RAM (GB)', price_col='Price ($)')
````

````python
# Compare RAM and GPU in different brands

# 1. Group data by GPU and RAM and calculate average price
gpu_ram_price = df.groupby(['GPU', 'RAM (GB)'])['Price ($)'].agg(['mean', 'count']).reset_index()
gpu_ram_price.columns = ['GPU', 'RAM (GB)', 'Mean Price ($)', 'Count']

# 2. Filter groups with less than 10 samples (to ensure significance)
gpu_ram_filtered = gpu_ram_price[gpu_ram_price['Count'] >= 10]

# 3. Convert data to matrix format (Pivot Table) for Heatmap
pivot_table = gpu_ram_filtered.pivot(index='RAM (GB)', columns='GPU', values='Mean Price ($)').fillna(0)

# 4. Plot Heatmap
plt.figure(figsize=(20, 12))
sns.heatmap(pivot_table,
annot=True,
fmt=".0f",
cmap='YlGnBu',
linewidths=0.5,
cbar_kws={'label': 'Average price ($)'}) # 'Average price ($)'

plt.title('Average price of laptops based on GPU and RAM', fontsize=14, pad=20) # 'Average price of laptops based on GPU and RAM'
plt.xlabel('GPU', fontsize=12)
plt.ylabel('RAM (GB)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

````python
# Display the number of samples per cell
pivot_count = gpu_ram_filtered.pivot(index='RAM (GB)', columns='GPU', values='Count').fillna(0)
plt.figure(figsize=(20, 12))
sns.heatmap(pivot_table, annot=pivot_count.values, fmt=".0f", cmap='YlGnBu')
````

````python
# TODO: Write a method to preprocess the storage space as described.
def parse_storage(storage_str):
    # Split the value and unit (e.g. '512GB SSD' → ['512', 'GB', 'SSD'])
    parts = storage_str.split()
    capacity_part = parts[0]
    storage_type = parts[1] if len(parts) > 1 else 'Unknown'

    # Extract the number and unit (e.g. '512GB' → 512 and 'GB')
    value = int(''.join(filter(str.isdigit, capacity_part)))
    unit = ''.join(filter(str.isalpha, capacity_part)).upper()

    # Convert to gigabytes
    if unit == 'TB':
        capacity_gb = value * 1024
    elif unit == 'GB':
        capacity_gb = value
    else:
        capacity_gb = None

    return capacity_gb, storage_type

# Apply the function to the column Storage
df[['Storage_Capacity_GB', 'Storage_Type']] = df['Storage'].apply(
lambda x: pd.Series(parse_storage(x))
)

# Show sample processed data
df[['Storage', 'Storage_Capacity_GB', 'Storage_Type']].head()
````

````python
from scipy.stats.mstats import winsorize

def cap_outliers(df: pd.DataFrame, column: str, limits: list[float]):
    """
    Caps outliers in a DataFrame column using winsorization.
    
    Args:
        df: Input pandas DataFrame
        column: Name of column to process
        limits: Tuple of lower and upper limits for winsorization (values between 0 and 1)
    
    Returns:
        DataFrame with outliers capped in the specified column
    """
    df = df.copy()  # Create a copy to avoid modifying original DataFrame
    
    # Calculate IQR bounds for informational purposes
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    print(f"{column}-->lower_bound:{ Q1 - 1.5 * IQR}---upper_bound{ Q3 + 1.5 * IQR}")

    # Apply winsorization to cap outliers
    df[column] = winsorize(df[column], limits=limits, inplace=False)
    
    return df

# Example usage: Winsorize top 5% outliers in 'Price ($)' column
df = cap_outliers(df=df, column='Price ($)', limits=[0, 0.05])
````

````python
from sklearn.preprocessing import MinMaxScaler

def MinMax_Scalery_y(y_train, y_val, y_test):
    """
    Applies MinMax scaling to target variables (y) across train, validation, and test sets.
    
    Args:
        y_train: Training target values (pandas Series)
        y_val: Validation target values (pandas Series)
        y_test: Test target values (pandas Series)
        
    Returns:
        Tuple of scaled y_train, y_val, y_test (pandas Series) with original indices preserved
    """
    
    # Initialize MinMaxScaler (normalizes data to [0, 1] range)
    scaler = MinMaxScaler()
    
    # Store original indices
    y_train_index = y_train.index
    y_val_index = y_val.index
    y_test_index = y_test.index
    
    # Create copies to avoid modifying originals
    y_train = y_train.copy()
    y_val = y_val.copy()
    y_test = y_test.copy()

    # Reshape and scale the data
    # Note: reshape(-1, 1) converts 1D array to 2D as required by sklearn
    y_train_scaled = scaler.fit_transform(y_train.values.reshape(-1, 1))  # Fit and transform on train
    y_val_scaled = scaler.transform(y_val.values.reshape(-1, 1))         # Transform validation
    y_test_scaled = scaler.transform(y_test.values.reshape(-1, 1))       # Transform test
    
    # Convert back to pandas Series with original indices and name
    y_train = pd.Series(y_train_scaled.flatten(), index=y_train_index, name="Price ($)")
    y_val = pd.Series(y_val_scaled.flatten(), index=y_val_index, name="Price ($)")
    y_test = pd.Series(y_test_scaled.flatten(), index=y_test_index, name="Price ($)")

    return y_train, y_val, y_test

# Scale target variables
y_train, y_val, y_test = MinMax_Scalery_y(y_train, y_val, y_test)
````

✅ VIF less than 5 → acceptable (lowest linear correlation with other features)

________________________________________________________________________________________________

````python
# Import variance_inflation_factor from statsmodels to detect multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Define a function to calculate VIF (Variance Inflation Factor) for all features
def calculate_vif(X):
    # Create an empty DataFrame to store results
    vif = pd.DataFrame()
    
    # Store feature names in the DataFrame
    vif["Feature"] = X.columns
    
    # Calculate VIF for each feature:
    # - X.values converts dataframe to numpy array
    # - i is the index of each feature column
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    
    return vif

# Calculate VIF for option 2 features:
# - Extract all column names from X_train_encoded
features_option2 = X_train_encoded.columns

# Compute VIF scores for these features
vif_result = calculate_vif(X_train_encoded[features_option2])

# Display the VIF results
print(vif_result)
````

````python
# TODO: If you need specific libraries, import them.
from sklearn.pipeline import Pipeline

from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor,XGBClassifier
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import learning_curve
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
````

````python
# Function to calculate model evaluation metrics
def evaluate_model(model, X, y, dataset_name):
    """Calculate model evaluation metrics including MAE, MSE, R² and RMSE"""
    # Model predictions on input data
    y_pred = model.predict(X)
    
    # Calculate and return various evaluation metrics
    return {
        f"{dataset_name} MAE": mean_absolute_error(y, y_pred),  # Mean Absolute Error
        f"{dataset_name} MSE": mean_squared_error(y, y_pred),   # Mean Squared Error
        f"{dataset_name} R²": r2_score(y, y_pred),             # R-squared
        f"{dataset_name} RMSE": np.sqrt(mean_squared_error(y, y_pred)),  # Root Mean Squared Error
    }

# Main function for training and evaluating machine learning models
def Machine_learning(models: dict, X_train_encoded, X_val_encoded, X_test_encoded, y_train, y_val, y_test):
    """
    Parameters:
    models: Dictionary of machine learning models
    X_*: Features for training, validation and test data
    y_*: Actual values for training, validation and test data
    
    Returns:
    results: Dictionary containing evaluation metrics for each model
    predictions: Dictionary containing predictions for each model
    """
    results = {}     # To store evaluation results
    predictions = {} # To store model predictions

    for name, pipeline in models.items():
        # Train model on training data
        pipeline.fit(X_train_encoded, y_train)

        # Store model predictions for different datasets
        predictions[name] = {
            "Train": pipeline.predict(X_train_encoded),
            "Validation": pipeline.predict(X_val_encoded),
            "Test": pipeline.predict(X_test_encoded)
        }

        # Evaluate model on different datasets
        train_metrics = evaluate_model(pipeline, X_train_encoded, y_train, "Train")
        val_metrics = evaluate_model(pipeline, X_val_encoded, y_val, "Validation")
        test_metrics = evaluate_model(pipeline, X_test_encoded, y_test, "Test")

        # Combine and store evaluation results
        results[name] = {**train_metrics, **val_metrics, **test_metrics}

    return results, predictions
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# TODO: Scatter plot of predicted vs actual values
# TODO: Visualize the relationship between actual and predicted values

def plot_pred_vs_actual(y_true, y_pred, title="Predicted vs Actual Values", figsize=(8, 6)):
    """
    Plots a scatter plot comparing actual vs predicted values with a reference line.
    
    Args:
        y_true: Array of actual target values
        y_pred: Array of predicted values
        title: Title for the plot
        figsize: Figure dimensions
    """
    plt.figure(figsize=figsize)
    sns.scatterplot(x=y_true, y=y_pred, alpha=0.6, color='blue', edgecolor='k')
    
    # Red dashed line representing perfect predictions
    max_val = max(np.max(y_true), np.max(y_pred)) * 1.05
    min_val = min(np.min(y_true), np.min(y_pred)) * 0.95
    plt.plot([min_val, max_val], [min_val, max_val], 'r--', lw=2, label='Ideal Line')
    
    plt.xlabel('Actual Values', fontsize=12)
    plt.ylabel('Predicted Values', fontsize=12)
    plt.title(title, fontsize=14)
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

def plot_all_pred_vs_actual(predictions, y_test):
    """
    Generates prediction vs actual plots for all models and datasets (Train, Val, Test).
    
    Args:
        predictions: Dictionary containing model predictions
        y_test: Test set actual values
    """
    for model_name, y_pred in predictions.items():
        # Training set plot
        plot_pred_vs_actual(y_train, y_pred["Train"], title=f"{model_name} - Train Predictions")
        # Validation set plot
        plot_pred_vs_actual(y_val, y_pred["Validation"], title=f"{model_name} - Validation Predictions")
        # Test set plot
        plot_pred_vs_actual(y_test, y_pred["Test"], title=f"{model_name} - Test Predictions")

def plot_error_distribution(y_true, y_pred, title="Error Distribution (Test Set)", figsize=(8, 6)):
    """
    Plots the distribution of prediction errors (residuals).
    
    Args:
        y_true: Array of actual target values
        y_pred: Array of predicted values
        title: Title for the plot
        figsize: Figure dimensions
    """
    errors = y_true - y_pred
    plt.figure(figsize=figsize)
    sns.histplot(errors, kde=True, color='blue', edgecolor='black', bins=30)
    plt.axvline(0, color='red', linestyle='--', label='Zero Error Line')
    plt.title(title, fontsize=14)
    plt.xlabel('Error', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.legend()
    plt.grid(alpha=0.3)
    plt.show()

# Generate prediction vs actual plots for all models
plot_all_pred_vs_actual(predictions, y_test)

# Generate error distribution plots for test set
for model_name, y_pred in predictions.items():
    plot_error_distribution(y_test, y_pred["Test"], title=f"{model_name} - Error Distribution (Test Set)")
````
